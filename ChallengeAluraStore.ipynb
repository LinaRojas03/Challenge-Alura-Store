## Descripcion del Proyecto


 A través del análisis de datos de ventas, rendimiento y reseñas de 4 tiendas de la cadena Alura Store; identificar la tienda menos eficiente

## Importación de Datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Ingreso Total

Calcular el ingreso total de cada tienda, sumando los valores de la columna **Precio** de cada conjunto de datos .

```python
# Calcular los ingresos totales de cada tienda sumando la columna 'Precio'
# Nota: 'Precio' es tipo: float64

ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Crear un diccionario con los resultados
ingresos = {
    'Tienda 1': ingreso_tienda1,
    'Tienda 2': ingreso_tienda2,
    'Tienda 3': ingreso_tienda3,
    'Tienda 4': ingreso_tienda4
}

# Convertir a DataFrame
df_ingresos = pd.DataFrame(list(ingresos.items()), columns=['Tienda', 'Ingreso Total'])

# Ordenar de forma ascendente los ingresos
df_ingresos = df_ingresos.sort_values(by='Ingreso Total').reset_index(drop=True)

# Mostrar resultado final
print("Ingresos totales por tienda (de menor a mayor):")
df_ingresos
```

### Gráfico de barras – Ingresos totales por tienda

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Usamos df_ingresos ya calculado antes
plt.figure(figsize=(8,5))
sns.barplot(data=df_ingresos, x='Tienda', y='Ingreso Total', palette='Blues_d')
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos ($ COP)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.ylim(900000000, 1200000000)   # Límite en eje Y
plt.show()

import warnings
warnings.filterwarnings('ignore')
```

## Ventas por Categoría
 Agrupar los datos por categoría y calcular el número de ventas de cada tipo, mostrando las categorías más vendidas.


```python
# Tienda 1
top_t1 = (
    tienda.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
    .head(3)
)
top_t1['Tienda'] = 'Tienda 1'

# Tienda 2
top_t2 = (
    tienda2.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
    .head(3)
)
top_t2['Tienda'] = 'Tienda 2'

# Tienda 3
top_t3 = (
    tienda3.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
    .head(3)
)
top_t3['Tienda'] = 'Tienda 3'

# Tienda 4
top_t4 = (
    tienda4.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
    .head(3)
)
top_t4['Tienda'] = 'Tienda 4'

# Unir en un solo DataFrame
df_top_cantidad = pd.concat([top_t1, top_t2, top_t3, top_t4], ignore_index=True)

# Mostrar resultado
df_top_cantidad
```

##  Productos más y menos vendidos

Identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda.

```python
# --- Tienda 1 ---
ventas_t1 = (
    tienda.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
)
mas_vendido_t1 = ventas_t1.head(1)
menos_vendido_t1 = ventas_t1.tail(1)

# --- Tienda 2 ---
ventas_t2 = (
    tienda2.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
)
mas_vendido_t2 = ventas_t2.head(1)
menos_vendido_t2 = ventas_t2.tail(1)

# --- Tienda 3 ---
ventas_t3 = (
    tienda3.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
)
mas_vendido_t3 = ventas_t3.head(1)
menos_vendido_t3 = ventas_t3.tail(1)

# --- Tienda 4 ---
ventas_t4 = (
    tienda4.groupby('Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=False)
)
mas_vendido_t4 = ventas_t4.head(1)
menos_vendido_t4 = ventas_t4.tail(1)

# Unir todo en un solo DataFrame para verlo junto
df_resumen = pd.concat([
    mas_vendido_t1.assign(Tienda='Tienda 1', Tipo='Más vendido'),
    menos_vendido_t1.assign(Tienda='Tienda 1', Tipo='Menos vendido'),
    mas_vendido_t2.assign(Tienda='Tienda 2', Tipo='Más vendido'),
    menos_vendido_t2.assign(Tienda='Tienda 2', Tipo='Menos vendido'),
    mas_vendido_t3.assign(Tienda='Tienda 3', Tipo='Más vendido'),
    menos_vendido_t3.assign(Tienda='Tienda 3', Tipo='Menos vendido'),
    mas_vendido_t4.assign(Tienda='Tienda 4', Tipo='Más vendido'),
    menos_vendido_t4.assign(Tienda='Tienda 4', Tipo='Menos vendido'),
], ignore_index=True)

df_resumen
```

### Gráfico de pastel – Distribución de categorías de productos en una tienda


```python
# Agrupar datos para cada tienda
categoria_t1 = tienda.groupby('Categoría del Producto').size()
categoria_t2 = tienda2.groupby('Categoría del Producto').size()
categoria_t3 = tienda3.groupby('Categoría del Producto').size()
categoria_t4 = tienda4.groupby('Categoría del Producto').size()

# Crear figura y ejes
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Tienda 1
categoria_t1.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                  colormap='Pastel1', ax=axs[0,0])
axs[0,0].set_title('Distribución de Categorías - Tienda 1')
axs[0,0].set_ylabel('')

# Tienda 2
categoria_t2.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                  colormap='Pastel2', ax=axs[0,1])
axs[0,1].set_title('Distribución de Categorías - Tienda 2')
axs[0,1].set_ylabel('')

# Tienda 3
categoria_t3.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                  colormap='Pastel1', ax=axs[1,0])
axs[1,0].set_title('Distribución de Categorías - Tienda 3')
axs[1,0].set_ylabel('')

# Tienda 4
categoria_t4.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                  colormap='Pastel2', ax=axs[1,1])
axs[1,1].set_title('Distribución de Categorías - Tienda 4')
axs[1,1].set_ylabel('')

plt.tight_layout()
plt.show()


```

## Calificación Promedio de la Tienda

Calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.


```python
# Calcular la calificación promedio de cada tienda
prom_t1 = tienda['Calificación'].mean()
prom_t2 = tienda2['Calificación'].mean()
prom_t3 = tienda3['Calificación'].mean()
prom_t4 = tienda4['Calificación'].mean()

# Crear DataFrame con los resultados
df_calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [prom_t1, prom_t2, prom_t3, prom_t4]
})

# Redondear a 2 decimales
df_calificaciones['Calificación Promedio'] = df_calificaciones['Calificación Promedio'].round(2)

df_calificaciones
```

## Envío Promedio por Tienda

Calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
# Calcular el costo de envío promedio para cada tienda
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

# Crear un DataFrame con los resultados
df_envio_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [envio_t1, envio_t2, envio_t3, envio_t4]
}).sort_values(by='Costo de Envío Promedio').reset_index(drop=True)

# Redondear a 2 decimales
df_envio_promedio['Costo de Envío Promedio'] = df_envio_promedio['Costo de Envío Promedio'].round(2)

df_envio_promedio
```

### Gráfico de dispersión – Relación entre calificación y costo de envío promedio
Para ver si hay algún patrón entre mejor calificación y costo de envío.

```python
# Usar el DataFrame df_calificaciones y df_envio_promedio combinados
df_merge = pd.merge(df_calificaciones, df_envio_promedio, on='Tienda')

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df_merge,
    x='Costo de Envío Promedio',
    y='Calificación Promedio',
    hue='Tienda',
    s=100,
    palette='deep'
)
plt.title('Relación entre Costo de Envío y Calificación Promedio')
plt.xlabel('Costo de Envío Promedio ($)')
plt.ylabel('Calificación Promedio')
plt.tight_layout()
plt.show()
```
